####1.Create file **yolov3_custom.cfg** with the same content as in yolov3.cfg

* change line batch to batch=64
* change line subdivisions to subdivisions=32
* change line classes=80 to your number of objects in each of 3 [yolo]-layers: classes = 5
  
* change [filters=255] to filters=(classes + 5)x3=30 in the 3 [convolutional] before each [yolo] layer
* change [burn_in=1000]  to 100
* change [max_baches] and [steps]


````python
%cp cfg/yolov3.cfg cfg/yolov3_custom.cfg

!sed -i 's/batch=1/batch=64/g' cfg/yolov3_custom.cfg
!sed -i 's/subdivisions=1/subdivisions=32/g' cfg/yolov3_custom.cfg
!sed -i 's/classes=80/classes=5/g' cfg/yolov3_custom.cfg
!sed -i 's/burn_in=1000/burn_in=100/g' cfg/yolov3_custom.cfg
!sed -i 's/filters=255/filters=30/g' cfg/yolov3_custom.cfg
!sed -i 's/max_batches = 505000/max_batches = 10000/g' cfg/yolov3_custom.cfg
!sed -i 's/steps=400000,450000/steps=8000,9000/g' cfg/yolov3_custom.cfg
#!sed -i 's/random=1/random=0/g' cfg/yolov3_custom.cfg
````

####5.Prepare Dataset for trainning

Used `OIDv4_ToolKit`


####6.Create file **train.txt** and **valid.txt** in directory `build\darknet\x64\data\` with filenames of your images with path relative to darknet


````python
import os
import numpy as np


trainset = open("build/darknet/x64/data/train.txt", "w")
validset = open("build/darknet/x64/data/valid.txt", "w")
listOfFiles = os.listdir('build/darknet/x64/data/train/obj/')   
for filename in listOfFiles:  
  if filename.endswith(".jpg"):
    if np.random.rand(1) < 0.8:
      trainset.write("build/darknet/x64/data/train/obj/"+filename+"\n")
    else:
      validset.write("build/darknet/x64/data/train/obj/"+filename+"\n")  
      
trainset.close()
validset.close()
````
